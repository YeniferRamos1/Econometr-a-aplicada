```python
df = df.set_index(['sid', 'year'])
# este es un identificador de estdo y de a√±o con el det.index puedes facilmente 
df
```

El primer modelo a regresionar es √∫nicamente con castle doctrine law (dcl) y control por a√±os (year)

```python
# Creamos year como una columna temporal para la regresi√≥n, puesto que al unsar el comando anterior la pasamos al √≠ndice
# y ya no e suna variable 
df['year'] = df.index.get_level_values('year')

# Funci√≥n para aplicar una regresi√≥n de efectos fijos
def run_did(y_var, data):
    formula = f"{y_var} ~  1 + cdl + C(year) + EntityEffects" # Formula de la regresi√≥n, entity effect son los efectos fijos por estado 
    model_1 = PanelOLS.from_formula(formula, data=df, weights=df['popwt']) # Aplicamos al conjunto de datos y usamos los pesos seg√∫n poblaci√≥n
    result_1 = model_1.fit(cov_type="clustered", cluster_entity=True)   # La clusterizaci√≥n por grupos cuando trabajemos a nivel de personas lo 
    #a nivel de personas o si hacemos a nivel de estad lo trbajamos anivel de estado . este m√©todo sirve para darnos errr√≥res estandar m√°s robustos 
    # y teneindo en cuanta que un individuo tiene correlaci√≥n con su pasado 
    print(result_1)
    return result_1,   

# Ahora, ejecutamos la funci√≥n para los elementos dentro de nuestras variables dependientes (yvar)
results_1 = {}
for y in yvar:
    results_1[y] = run_did(y, df)
```

```python
# interancci√≥n entre zona y a√±o si cumple estos requisitos se coloca 1 
# Creamos year como una columna temporal para la regresi√≥n
df['year'] = df.index.get_level_values('year')

# Convertimos la lista de regiones en una cadena separada por '+' para poder usarla en la regresi√≥n
region_regress = " + ".join(region)

# Funci√≥n para aplicar una regresi√≥n de efectos fijos
def run_did(y_var, data):
    formula = f"{y_var} ~ 1 + cdl + C(year) + {region_regress} + EntityEffects" # Formula de la regresi√≥n
    model_2= PanelOLS.from_formula(formula, data=df, weights=df['popwt'],  check_rank=False, drop_absorbed=True) # Aplicamos al conjunto de datos y usamos los pesos seg√∫n poblaci√≥n
    result_2 = model_2.fit(cov_type= "clustered", cluster_entity=True,)   # La clusterizaci√≥n por grupos
    print(result_2)
    return result_2, 

# Ahora, ejecutamos la funci√≥n para los elementos dentro de nuestras variables dependientes (yvar)
results_2 = {}
for y in yvar:
    results_2[y] = run_did(y, df)
```
